### Data preprocessing

```python
def resize(input_image, input_mask):
    input_image = tf.image.resize(input_image, (128, 128), method="nearest")
    input_mask = tf.image.resize(input_mask, (128, 128), method="nearest")

    return input_image, input_mask
```

```python
def augment(input_image, input_mask):
    if tf.random.uniform(()) > 0.5:
        # random LR flipping of image and mask
        input_image = tf.image.flip_left_right(input_image)
        input_mask = tf.image.flip_left_right(input_mask)

    return input_image, input_mask
```

```python
def normalize(input_image, input_mask):
    input_image = tf.cast(input_image, tf.float32) / 255.0
    input_mask -= 1

    return input_image, input_mask
```

```python
def load_image_train(datapoint):
    input_image = datapoint["image"]
    input_mask = datapoint["segmentation_mask"]
    input_image, input_mask = resize(input_image, input_mask)
    input_image, input_mask = augment(input_image, input_mask)
    input_image, input_mask = normalize(input_image, input_mask)

    return input_image, input_mask
```

```python
def load_image_test(datapoint):
    input_image = datapoint["image"]
    input_mask = datapoint["segmentation_mask"]
    input_image, input_mask = resize(input_image, input_mask)
    input_image, input_mask = normalize(input_image, input_mask)

    return input_image, input_mask
```

```python
train_dataset = dataset["train"].map(load_image_train, num_parallel_calls=tf.data.AUTOTUNE)
test_dataset = dataset["test"].map(load_image_train, num_parallel_calls=tf.data.AUTOTUNE)
```

```python
print(train_dataset)
```

Create train, validation and test batches. Original test dataset is split into validation and test batches.

```python
BATCH_SIZE=64
BUFFER_SIZE=1000
```

```python
train_batches = train_dataset.cache().shuffle(BUFFER_SIZE).batch(BATCH_SIZE).repeat()
train_batches = train_batches.prefetch(buffer_size=tf.data.experimental.AUTOTUNE)
validation_batches = test_dataset.take(3000).batch(BATCH_SIZE)
test_batches = test_dataset.skip(3000).take(669).batch(BATCH_SIZE)
```

## U-Net building blocks

```python
def double_conv_block(x, n_filters):
    # Conv2d then ReLU activation
    x = layers.Conv2D(
        filters=n_filters,
        kernel_size=3,
        padding="same",
        activation="relu",
        kernel_initializer="he_normal"
    )(x)
    # Conv2d then ReLU activation
    x = layers.Conv2D(
        filters=n_filters,
        kernel_size=3,
        padding="same",
        activation="relu",
        kernel_initializer="he_normal"
    )(x)

    return x
```

```python
def downsample_block(x, n_filters):
    f = double_conv_block(x, n_filters)
    p = layers.MaxPool2D(pool_size=2)(f)
    p = layers.Dropout(rate=0.3)(p)

    return f, p
```

```python
def upsample_block(x, conv_features, n_filters):
    # upsample
    x = layers.Conv2DTranspose(
        filters=n_filters,
        kernel_size=3,
        strides=2,
        padding="same"
    )(x)
    # concatenate
    x = layers.concatenate([x, conv_features])
    # dropout
    x = layers.Dropout(rate=0.3)(x)
    # Conv2D twice with ReLU activation
    x = double_conv_block(x, n_filters)

    return x
```

## Build the U-Net model

```python
def build_unet_model():
    # inputs
    inputs = layers.Input(shape=(128, 128, 3))

    # encoder: contracting path - downsample
    # block 1 - downsample
    f1, p1 = downsample_block(inputs, 64)
    # block 2 - downsample
    f2, p2 = downsample_block(p1, 128)
    # block 3 - downsample
    f3, p3 = downsample_block(p2, 256)
    # block 4 - downsample
    f4, p4 = downsample_block(p3, 512)

    # block 5 - bottleneck
    bottleneck = double_conv_block(p4, 1024)

    # decoder: expanding path - upsample
    # block 6 - upsample
    u6 = upsample_block(bottleneck, f4, 512)
    # block 7 - upsample
    u7 = upsample_block(u6, f3, 256)
    # block 8 - upsample
    u8 = upsample_block(u7, f2, 128)
    # block 9 - upsample
    u9 = upsample_block(u8, f1, 64)

    # outputs
    outputs = layers.Conv2D(
        filters=3,
        kernel_size=1,
        padding="same",
        activation="softmax"
    )(u9)

    # unet model with Keras functional API
    unet_model = tf.keras.Model(inputs, outputs, name="U-Net")

    return unet_model
```

```python
unet_model = build_unet_model()
```

## Compile and train U-Net

```python
unet_model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
# config
NUM_EPOCHS = 20

TRAIN_LENGTH = info.splits["train"].num_examples
STEPS_PER_EPOCH = TRAIN_LENGTH // BATCH_SIZE

VAL_SUBSPLITS = 5
TEST_LENGTH = info.splits["test"].num_examples
VALIDATION_STEPS = TEST_LENGTH // BATCH_SIZE // VAL_SUBSPLITS

# train network
model_history = unet_model.fit(
    train_batches,
    epochs=NUM_EPOCHS,
    steps_per_epoch=STEPS_PER_EPOCH,
    validation_data=validation_batches,
    validation_steps=VALIDATION_STEPS,
)
```

## Learning curve from model history

```python
def display_learning_curves(history):
    acc = history.history["accuracy"]
    val_acc = history.history["val_accuracy"]

    loss = history.history["loss"]
    val_loss = history.history["val_loss"]

    epochs_range = range(NUM_EPOCHS)

    fig = plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label="train accuracy")
    plt.plot(epochs_range, val_acc, label="validation accuracy")
    plt.title("Accuracy")
    plt.xlabel("Epoch #")
    plt.ylabel("Accuracy")
    plt.grid()
    plt.legend(loc="lower right")

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label="train loss")
    plt.plot(epochs_range, val_loss, label="validation loss")
    plt.title("Loss")
    plt.xlabel("Epoch #")
    plt.ylabel("Loss")
    plt.grid()
    plt.legend(loc="upper right")

    fig.tight_layout()
    plt.show()
```

```python
display_learning_curves(model_history)
```

## Predictions with U-net model

```python
def create_mask(pred_mask):
    pred_mask = tf.argmax(pred_mask, axis=-1)
    pred_mask = pred_mask[..., tf.newaxis]
    return pred_mask[0]
```

```python
def show_predictions(dataset=None, num=1):
    if dataset is not None:
        for image, mask in dataset.take(num):
            pred_mask = unet_model.predict(image)
            display([image[0], mask[0], create_mask(pred_mask)])
    else:
        pred_mask = unet_model.predict(sample_image[tf.newaxis, ...])
        display([sample_image, sample_mask, create_mask(pred_mask)])
        
```
